```python
normal_data_features_matrix = df[df['Class'] == 0]
```

```python
# we now have to drop the class columns
normal_data_features_matrix = normal_data_features_matrix.drop(columns=['Class', 'Time'])
# then we convert back to numpy
X = normal_data_features_matrix.to_numpy()
```

```python
X
```

```python
# center X features matrix
means = np.mean(X, axis=0)
X_center = X - means
```

```python
X_center
```

```python
# now we perform a SVD decomposition on the centered X matrix
U, S, V = scipy.linalg.svd(X_center, full_matrices=False)
```

```python
# now we can compute the eigenvalues
n = X.shape[0]
Î» = S ** 2 / (n - 1)
```

```python
cumvar = np.cumsum(Î»)
percentvar = cumvar / cumvar[-1]
```

```python
percentvar
```

```python
fig, ax = plt.subplots()
ax.plot(np.arange(len(Î»)), percentvar)
ax.set(xlabel='Number of features', ylabel='Explained Variance',
       title='Explained Variance as a function of features left')
ax.grid()
fig.savefig("variance_wrt_nb_features.png")
plt.show()
```

```python
# By the elbow method we should choose k = 13
k2 = 2
X2 = U[:, :k2] @ np.diag(S)[:k2, :k2]
X2
```

```python
fig, ax = plt.subplots()
x = X2[:, 0]
y = X2[:, 1]

ax.scatter(x, y)
ax.set(title='Data after PCA reduction to 2 dimensions')
fig.savefig("2dPCA.png")
plt.show()
```

```python
# Now we do the same for 3 dimensional data
k3 = 3
X3 = U[:, :k3] @ np.diag(S)[:k3, :k3]

fig, ax = plt.subplots()
ax = fig.add_subplot(projection='3d')

ax.scatter(X3[:, 0], X3[:, 1], X3[:, 2])

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
fig.savefig('3dPCA.png')
plt.show()
```

```python
def em(dataset, classes, n_iterations, random_seed):
    n_samples = dataset.shape[0]
    n_features = dataset.shape[1]
    dataset = torch.tensor(dataset)
    torch.manual_seed(random_seed)

    # initial guesses for the parameters
    class_probs = torch.rand(n_classes)
    class_probs /= class_probs.sum()
    mus = torch.tensor(kmeans.cluster_centers_)
    sigmas = torch.eye(n_features) * int(1e3)
    sigmas = sigmas.repeat(n_classes, 1, 1)

    # now we iterate
    for em_iter in tqdm.tqdm(range(n_iterations)):
        # E step
        responsibilities = torch.empty((n_samples, n_classes))
        dists = [D.MultivariateNormal(mus[i], sigmas[i]) for i in range(n_classes)]
        s = 0
        for i in range(n_classes):
            inter = torch.exp(dists[i].log_prob(dataset)) * class_probs[i]
            responsibilities[:, i] = inter
            s += torch.unsqueeze(inter, -1)
        print(s)
        responsibilities / s

        # M step

        # Calculate the new class probs
        new_class_probs = torch.sum(responsibilities, dim=0)
        # Calculate the new class means
        weighted_dataset = dataset.unsqueeze(1) * responsibilities.unsqueeze(-1)
        weighted_sums = torch.sum(weighted_dataset, dim=0)
        new_mus = weighted_sums / new_class_probs.unsqueeze(-1)
        # Calculate the new class covariance matrices
        new_sigmas = torch.empty_like(sigmas)
        for i in range(n_classes):
            new_mu_i = new_mus[i]
            recentered_dataset = dataset - new_mu_i
            responsibilities_class_i = responsibilities[:, i].view(-1, 1, 1)
            outer_products = torch.einsum('ij,ik->ijk', recentered_dataset, recentered_dataset)
            weighted_outer_products = outer_products * responsibilities_class_i
            sums_outer_products = torch.sum(weighted_outer_products, axis=0)
            cov_matrix_i = sums_outer_products / new_class_probs[i]
            new_sigmas[i] = cov_matrix_i

        # Reinstantiate class_probs, mus and sigmas
        class_probs = new_class_probs / n_samples
        mus = new_mus
        sigmas = new_sigmas
        print(sigmas)
        

    # return the parameters
    return class_probs, mus, sigmas
        

n_classes = 2
pis_prob, mus_, sigmas_ = em(X3, n_classes, n_iterations, random_seed)
```

```python
# credit for the plot code : https://github.com/ChenTaHung/GMM-Visualization/blob/main/src/main/GMMViz/GmmPlot.py
from scipy.stats import chi2
from scipy.special import erf
from mpl_toolkits.mplot3d import Axes3D
def eigendecomposition(A):
    Î›, V = scipy.linalg.eig(A)
    return Î›, V
def plot_3d_gaussian(mu, sigma, ax, color, n_std=2, alpha = 0.4):
    eigenvalues, eigenvectors = eigendecomposition(sigma)
    eigenvalues, eigenvectors = eigenvalues.astype(np.float32), eigenvectors.astype(np.float32)
    # Adjust radii by the chi-squared distribution quantile for confidence interval
    radii = np.sqrt(eigenvalues) * n_std * np.sqrt(chi2.ppf((1 + erf(n_std / np.sqrt(2))) / 2, df=2))
    # Generate a grid of points representing the ellipsoid
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0, np.pi, 100)
    x = radii[0] * np.outer(np.sin(v), np.cos(u))
    y = radii[1] * np.outer(np.sin(v), np.sin(u))
    z = radii[2] * np.outer(np.cos(v), np.ones_like(u))

     # Rotate the points according to the eigenvectors
    ellipsoid = np.dstack((x, y, z))
    
    ellipsoid = ellipsoid @ eigenvectors + mu
    # Plot the surface
    ax.plot_surface(ellipsoid[:,:,0], ellipsoid[:,:,1], ellipsoid[:,:,2], color=color, alpha=alpha)
```

```python
# For comparison let's now plot all of the Gaussians together
fig, ax = plt.subplots()
ax = fig.add_subplot(projection='3d')

ax.scatter(X3[:, 0], X3[:, 1], X3[:, 2])
colors = ['red', 'blue']

for i in range(n_classes):
    print(mus_[i], sigmas_[i])
    plot_3d_gaussian(mus_[i], sigmas_[i], ax=ax, color=colors[i])

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

fig.savefig('GMM_fit_to_data.png')
plt.show()
```
